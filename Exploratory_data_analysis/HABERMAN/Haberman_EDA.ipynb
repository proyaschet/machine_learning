```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
column_names = ['Age', 'Year operation', 'Axillary nodes detected', 'Survival status']
data = pd.read_csv("haberman.csv", names = column_names)
```

```python
data['Survival status'] = data['Survival status'].replace(1,'>5 years')
data['Survival status'] = data['Survival status'].replace(2,'<5 years')
```

```python
data.head()
```

In Survival Status 1 represents the patient survived 5 years or longer  and 2 represents the patient died within 5 year

```python
counts, bin_edges = np.histogram(data['Age'], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf) # SUM OF PDF OR INTEGRATION OF PDF = CDF AND DIFFERENTIATION OF CDF = PDF
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Age')
plt.show()
```

```python
counts, bin_edges = np.histogram(data['Year operation'], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf) # SUM OF PDF OR INTEGRATION OF PDF = CDF AND DIFFERENTIATION OF CDF = PDF
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Year operation')
plt.show()
```

```python
counts, bin_edges = np.histogram(data['Axillary nodes detected'], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf) # SUM OF PDF OR INTEGRATION OF PDF = CDF AND DIFFERENTIATION OF CDF = PDF
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Axillary nodes detected')
plt.show()
```
